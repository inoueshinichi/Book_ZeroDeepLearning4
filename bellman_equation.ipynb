### ベルマン方程式
+ 報酬$R_{t+1}$はt+1で取得する作法

#### 状態価値関数のベルマン方程式

$$V(s) = E[R_{t+1} + \gamma V(s') | S_t = s]$$

$$V(s) = \sum_{a} \pi(a|s) \sum_{s', r} p(s'| s, a) [r + \gamma V(s')]$$

#### 行動価値関数のベルマン方程式

$$Q(s, a) = E[R_{t+1} + \gamma Q(s', a') | S_t = s, A_t = a]$$

$$Q(s, a) = \sum_{s', r} p(s'| s, a) [r + \gamma \sum_{a'} \pi(a'|s') Q(s', a')]$$

#### 最適ベルマン方程式

$$V^*(s) = \max_{a} \sum_{s', r} p(s'| s, a) [r + \gamma V^*(s')]$$

$$Q^*(s, a) = \sum_{s', r} p(s'| s, a) [r + \gamma \max_{a'} Q^*(s', a')]$$
